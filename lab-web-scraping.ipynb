```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin

# Función para convertir la calificación por estrellas en un valor numérico
def get_star_rating(rating_str):
    ratings_dict = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
    return ratings_dict.get(rating_str, 0)

# Función para extraer los datos de libros
def scrape_books(min_rating=4, max_price=20):
    base_url = "http://books.toscrape.com/catalogue/page-{}.html"
    book_data = []
    page_num = 1
    has_next_page = True
    site_url = "http://books.toscrape.com/"

    while has_next_page:
        response = requests.get(base_url.format(page_num))
        soup = BeautifulSoup(response.content, 'html.parser')

        # Encontrar todos los libros en la página actual
        books = soup.find_all('article', class_='product_pod')

        for book in books:
            # Extraer el título
            title = book.h3.a['title']
            
            # Extraer el precio (por ejemplo, '£53.23')
            price_str = book.find('p', class_='price_color').text.strip()
            price = float(price_str[1:])  # Eliminar '£' y convertir a float
            
            # Extraer la calificación
            rating_class = book.find('p', class_='star-rating')['class'][1]  # La segunda clase contiene la calificación
            rating = get_star_rating(rating_class)

            # Extraer la disponibilidad
            availability = book.find('p', class_='instock availability').text.strip()
            
            # Verificar si el libro cumple con los criterios de calificación y precio
            if rating >= min_rating and price <= max_price:
                # Obtener el href del enlace
                relative_book_url = book.h3.a['href']
                # Corregir la URL, añadiendo "catalogue/" si falta
                if not relative_book_url.startswith('catalogue'):
                    relative_book_url = "catalogue/" + relative_book_url
                # Construir la URL completa
                book_url = urljoin(site_url, relative_book_url)
                
                # Visitar la página individual del libro para obtener más detalles
                book_page = requests.get(book_url)
                book_soup = BeautifulSoup(book_page.content, 'html.parser')

                # Extraer el género desde la barra de navegación (breadcrumb)
                genre_tag = book_soup.find('ul', class_='breadcrumb')
                if genre_tag and len(genre_tag.find_all('li')) > 2:
                    genre = genre_tag.find_all('li')[2].text.strip()  # El género está en el tercer elemento del breadcrumb
                else:
                    genre = 'Unknown'

                # Extraer el UPC desde la tabla de especificaciones del producto
                product_info_table = book_soup.find('table', class_='table table-striped')
                upc = 'N/A'
                if product_info_table:
                    rows = product_info_table.find_all('tr')
                    for row in rows:
                        header = row.find('th').text.strip()
                        if header == 'UPC':
                            upc = row.find('td').text.strip()
                            break

                # Extraer la descripción si existe
                description_tag = book_soup.find('div', id='product_description')
                description = 'No description available.'
                if description_tag:
                    next_sibling = description_tag.find_next_sibling('p')
                    if next_sibling:
                        description = next_sibling.text.strip()

                # Agregar los datos del libro a la lista
                book_data.append({
                    'UPC': upc,
                    'Title': title,
                    'Price (£)': price,
                    'Rating': rating,
                    'Genre': genre,
                    'Availability': availability,
                    'Description': description
                })

        # Verificar si hay una siguiente página
        has_next_page = bool(soup.find('li', class_='next'))
        page_num += 1

    # Convertir la lista a un DataFrame de pandas
    df = pd.DataFrame(book_data)
    return df
```

```python
# Extraer libros con una calificación mínima de 4 y precio máximo de £20
df_books = scrape_books(min_rating=4, max_price=20)
    
# Mostrar las primeras filas del DataFrame
print(df_books.head())
    
# Guardar los datos en un archivo CSV
df_books.to_csv("books_to_scrape.csv", index=False)
```
